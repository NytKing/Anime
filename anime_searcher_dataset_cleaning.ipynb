```python
#importing all the necessary libraries and data
import pandas as pd
import numpy as np

anime = pd.read_csv("https://raw.githubusercontent.com/NytKing/Anime/main/animes.csv")
```

```python
#just viewing the data
anime.head(5)
```

Now, if I directly import the excel file to sql server, it does not get uploaded, this is because that there are columns which has charcters other than words and numbers, that is why I have uploaded the data to first remove all the non-numeric characters and alphabets.


So, I have taken 4 four columns and removed the non-numeric characters and alphabets 

```python
anime['aired']=anime['aired'].str.replace(r'\W',' ')
anime['title']=anime['title'].str.replace(r'\W',' ')
anime['synopsis']=anime['synopsis'].str.replace(r'\W',' ')
anime['genre']=anime['genre'].str.replace(r'\W',' ')
```

Now, if we see the output, we see that the non-numeric characters and alphabets are removed.
